# 당진태양고도 크롤링

```python
# angle_1 = 방위각 , _2 = 고도 , _3 = 적경, _4 = 적위
```

```python
Chrome_directory = 'D:/02.users/wodud2468/chromedriver/chromedriver.exe'
driver = webdriver.Chrome(Chrome_directory)
options = webdriver.ChromeOptions()
date = []
hour = []
angle_1 = []
angle_2 = []
angle_3 = []
angle_4 = []
for d in day.astype(str):
    url = 'https://astro.kasi.re.kr/life/pageView/10?useElevation=1&lat=37.05127236188373&lng=126.51409076261092&elevation=-117.96642423355715&output_range=1&date={}&hour=13&minute=30&second=&address=%EC%B6%A9%EC%B2%AD%EB%82%A8%EB%8F%84+%EB%8B%B9%EC%A7%84%EC%8B%9C+%EC%84%9D%EB%AC%B8%EB%A9%B4+%EA%B5%90%EB%A1%9C%EA%B8%B8+30#'.format(d)
    driver.get(url)
    for i in range(1,25):
        date.append(d)
        hour.append(driver.find_element_by_xpath('//*[@id="sun-height-table"]/table/tbody/tr[{}]/td[1]'.format(i)).text)
        angle_1.append(driver.find_element_by_xpath('//*[@id="sun-height-table"]/table/tbody/tr[{}]/td[2]'.format(i)).text)
        angle_2.append(driver.find_element_by_xpath('//*[@id="sun-height-table"]/table/tbody/tr[{}]/td[3]'.format(i)).text)
        angle_3.append(driver.find_element_by_xpath('//*[@id="sun-height-table"]/table/tbody/tr[{}]/td[4]'.format(i)).text)
        angle_4.append(driver.find_element_by_xpath('//*[@id="sun-height-table"]/table/tbody/tr[{}]/td[5]'.format(i)).text)
```

```python
df = pd.DataFrame([date,hour,angle_1,angle_2,angle_3,angle_4]).T
df.columns = ['date','hour','방위각','고도','적경','적위']
```

```python
# ./data/태양광/ 에 크롤링 완료된 데이터가 있습니다.
df.to_csv('dangjin_sun_height.csv')
```
